```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
###################################################
figure_width = 12
###################################################

sns.set_context('paper', font_scale=3)
sns.set_style('ticks')
plt.rcParams['figure.figsize'] = (figure_width, 10)
out_path = 'img/'

###################################################
savefigs = True
###################################################
```

```python
prefix = 'P1'

tanh = pd.read_csv('datos/run_Adam_0_001_tanh-tag-loss.csv')
tanh.columns = tanh.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
relu = pd.read_csv('datos/run_Adam_0_001_relu-tag-loss.csv')
relu.columns = relu.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
sigmoid = pd.read_csv('datos/run_Adam_0_001_sigmoid-tag-loss.csv')
sigmoid.columns = sigmoid.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
leaky = pd.read_csv('datos/run_Adam_0_001_leaky_relu-tag-loss.csv')
leaky.columns = leaky.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
plt.plot(tanh.step, tanh.value, label='tanh', lw= 4)
plt.plot(sigmoid.step, sigmoid.value, label='sigmoid', lw= 4)
plt.plot(relu.step, relu.value, label='relu', lw= 4)
plt.plot(leaky.step, leaky.value, label='leaky relu', lw= 4)
plt.legend()
plt.xlabel('Step')
plt.ylabel('Loss')
plt.title('Comparación de función de loss para \n distintas funciones de activación')
plt.grid()
plt.ylim([0, 0.5])
plt.savefig(out_path + prefix + 'funciones.pdf', bbox_inches='tight')
```

```python
plt.plot(relu.step[-25:], relu.value[-25:], label='relu', lw= 4)
plt.plot(leaky.step[-25:], leaky.value[-25:], label='leaky relu', lw= 4)
plt.legend()
plt.xlabel('Step')
plt.ylabel('Loss')
plt.title('Comparación de función de loss para \n distintas funciones de activación')
plt.grid()
plt.savefig(out_path + prefix + 'relu_vs_leaky.pdf', bbox_inches='tight')
```

```python
prefix = 'P2'
relu_01 = pd.read_csv('datos/run_Adam_0_01_relu-tag-loss.csv')
relu_01.columns = relu_01.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
plt.plot(relu.step, relu.value, label='Learning rate = 0.001', lw= 4)
plt.plot(relu_01.step, relu_01.value, label='Learning rate = 0.01', lw= 4)
plt.legend()
plt.xlabel('Step')
plt.ylabel('Loss')
plt.title('Comparación de Adam con Relu y\ndistintas tasas de aprendizaje')
plt.grid()
plt.ylim([0, 0.3])
plt.savefig(out_path + prefix + 'OptimizadorMNIST.pdf', bbox_inches='tight')
```
